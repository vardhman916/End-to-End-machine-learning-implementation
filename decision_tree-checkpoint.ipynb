```python
import pandas as pd
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
df = load_breast_cancer()
df
```

```python
data = pd.DataFrame(df.data)
data
```

```python
data.columns = df.feature_names
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
X = data
y = df.target
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler# the use of standardisation to make data in same scale 
##otherwise our  model is focus on higher values 
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
decision_tree = DecisionTreeClassifier()
decision_tree
```

```python
parameter = {'criterion':['gini','entropy','log_loss'],'splitter':['best','random'],'max_depth':[3,4]}
decision_tree = GridSearchCV(decision_tree,parameter,scoring = 'accuracy',cv = 5)
decision_tree = decision_tree.fit(X_train,y_train)
decision_tree
```

```python
print("Best parameters:", decision_tree.best_params_)
```

```python
prediction = decision_tree.predict(X_test)
prediction
```

```python
decision_tree.predict_proba(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,prediction)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,prediction)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,prediction,target_names = ['malignant', 'benign']))
```

```python
from matplotlib import pyplot as plt
from sklearn import tree
best_model = decision_tree.best_estimator_

# Assuming 'treemodel' is a trained decision tree model
plt.figure(figsize=(15, 10))

# Plot the tree
tree.plot_tree(best_model, filled=True)

# Show the plot
plt.show()
```
