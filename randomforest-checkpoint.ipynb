```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('tested.csv')
data
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data['Age'] = data['Age'].fillna(data['Age'].mode()[0])
data['Cabin'] = data['Cabin'].fillna(data['Cabin'].mode()[0])
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace = True)
```

```python
data.dtypes # it wil return series with column name as index and value as dattype
```

```python
# getting dummiesfrom all other categorical variable
for i in data.dtypes[data.dtypes == 'object'].index: # it will give object column
    for_dummy = data.pop(i)
    dummies = pd.get_dummies(for_dummy,prefix = i).astype(int)
    data = pd.concat([data,dummies],axis = 1)
y = data.pop('Survived')
X = data
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler# the use of standardisation to make data in same scale 
##otherwise our  model is focus on higher values 
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
randomforest = RandomForestClassifier()
randomforest
```

```python
parameter = {'criterion':['gini', 'entropy', 'log_loss'],'max_features':['sqrt', 'log2', None],'class_weight':['balanced', 'balanced_subsample'],'max_depth':[3,4]}
randomforest = GridSearchCV(randomforest,parameter,scoring = 'accuracy',cv = 5)
randomforest = randomforest.fit(X_train,y_train)
randomforest
```

```python
print("Best parameters:", randomforest.best_params_)
```

```python
prediction = randomforest.predict(X_test)
prediction
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,prediction)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,prediction))
```
