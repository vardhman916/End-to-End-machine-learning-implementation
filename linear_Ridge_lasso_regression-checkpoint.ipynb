```python
import pandas as pd ##use for to clean the data
import matplotlib.pyplot as plt ## it is use to visualise data 
import seaborn as sns ## it is use to visualise data 
import numpy as np ##It has a vast  a collection of mathematical function to opertae the data
from sklearn.datasets import fetch_california_housing



#sklearn is use to implement machine learning alogoritm like performanc metric,preprocessing(standard scaler)
#model selection
```

```python
# Clear the cache where the dataset is stored
data = fetch_california_housing()
data
```

```python
X = df
y = data.target
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=42)
'''If you set random_state=42, every time you select students for the study 
and test groups, it will always choose the same 80 and 20 students, ensuring consistency.'''
y_test
```

```python
X_train
y_test
```

```python
from sklearn.preprocessing import StandardScaler# the use of standardisation to make data in same scale 
##otherwise our  model is focus on higher values 
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
'''(x-m)/s
When using .fit_transform() on the training data, the model calculates 
the mean and standard deviation of the training data and then standardizes it (scales it).
When using .transform() on the test data, the model uses the same mean and standard deviation 
that were calculated from the training data to standardize the test data, 
ensuring consistency and preventing information leakage'''

#scaler.inverse_transform(X_train) by the help of it we will get earlier data befor standardisation
```

```python
#what is hyper parameter
'''he goal of hyperparameter tuning is to find the combination of hyperparameters that leads to 
the best model performance. This is typically done through methods like'''
```

```python
#linear regression
from sklearn.linear_model import LinearRegression
## cross validation
from sklearn.model_selection import cross_val_score
'''cross_val_score is a function in scikit-learn that performs 
cross-validation on a model by splitting the data into multiple folds, training the
model on some folds, and testing it on others, then returns an array of performance scores'''
```

```python
regression = LinearRegression()
regression.fit(X_train,y_train)
```

```python
mse  = cross_val_score(regression,X_train,y_train,scoring = 'neg_mean_squared_error',cv = 10)
#Scoring refers to the function  to evaluate the performance of a model. 
# cv = 6 means divide in 6 fold and and evaluted by 6 time 
```

```python
np.mean(mse)
```

```python
#regression
reg_pred = regression.predict(X_test)
```

```python
reg_pred
```

```python
## Now we will compare with the truth value
import seaborn as sns
sns.displot(reg_pred-y_test,kind = 'kde')
```

```python
from sklearn.metrics import r2_score
score = r2_score(reg_pred,y_test)
score
#it should be less
```

```python
# Reshape predictions to 2D (required by inverse_transform)
y_pred_scaled_reshaped  = reg_pred.reshape(-1, 8)

# Use the scaler to inverse transform the scaled predictions back to the original scale
y_pred_original = scaler.inverse_transform(y_pred_scaled_reshaped)
# Flatten the predictions to 1D
y_pred_original_flat = y_pred_original.flatten()  # Convert to 1D

# Now you can compare y_test (1D) with y_pred_original_flat (1D)
y_pred_original_flat
```

<h1> Ridge <h6>

```python
#why we use ridge 
# to hndle the overfitting 
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
'''why we use grid searchcv becuase in ridge regression we use alfa parameter so to search the best hyperparameter
gridsearch cv make all the possible combination and find best paramter'''
```

```python

ridge_regression = Ridge()
ridge_regression
```

```python
parameter = {'alpha':[1,2,5,5.5,6,6.5,7,7.5,8,10,20,30,40,50,55,60,70,80,90]}
ridgeCV = GridSearchCV(ridge_regression,parameter,scoring = 'neg_mean_squared_error',cv = 5)
ridgeCV.fit(X_train,y_train)
```

```python
print(ridgeCV.best_params_)
```

```python
print(ridgeCV.best_score_) # best for value near the 1
```

```python
rid_pred = ridgeCV.predict(X_test)
score1 = r2_score(rid_pred,y_test)
score1
```

```python
sns.displot(rid_pred-y_test,kind = 'kde')
```
